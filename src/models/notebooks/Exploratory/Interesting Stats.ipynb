```python
def convert_height_to_cm(height):
    if height=='--':
        return 0
    feet=''
    inches=''
    while height[0]!="'":
        feet+=height[0]
        height=height[1:]
    height=height[2:]
    while height[0]!='"':
        inches+=height[0]
        height=height[1:]
    return int(feet)*12*2.54 + int(inches)*2.54

def convert_reach_to_cm(reach):
    if reach=='--':
        return 0
    return int(reach[0:2])*2.54

def top_5(stat, max_min = 'max'):
    result={}
    column = ufcfighterscrap[stat]
    while len(set(result.keys()))<5:
        if len(column)==0:
            break
        if max_min == 'max':
            max_value = column.max()
            max_index = column.idxmax()
            if max_value in result.keys():
                result[max_value].append(ufcfighterscrap['name'][max_index])
            else:
                result[max_value]=[ufcfighterscrap['name'][max_index]]
            column = column.drop(max_index)
        elif max_min == 'min':
            min_value = column.min()
            min_index = column.idxmin()
            if min_value in result.keys():
                result[min_value].append(ufcfighterscrap['name'][min_index])
            else:
                result[min_value]=[ufcfighterscrap['name'][min_index]]
            column = column.drop(min_index)
    return result
```

```python
#making list of top 5 and bottom 5 scores for each good statistic
fighters_with_top_5_stats={}
fighters_with_bottom_5_stats={}
for stat in good_statistics:
    print(stat)
    fighters_with_top_5_stats[stat]=top_5(stat,'max')
    fighters_with_bottom_5_stats[stat]=top_5(stat,'min')
    
```

```python
def find_interesting_stats(fighter):
    interesting_stats = {}
    for stat in fighters_with_top_5_stats.keys():
        #we count how many fighters rank higher
        count=0
        below=[]
        key_number=1
        for key in fighters_with_top_5_stats[stat].keys():
            #print(stat,key,fighters_with_top_5_stats[stat][key])
            if fighter in fighters_with_top_5_stats[stat][key]:
                tied_with = fighters_with_top_5_stats[stat][key].copy()
                tied_with.remove(fighter)
                if len(fighters_with_top_5_stats[stat][key])<250:
                    interesting_stats[stat]={'value': key,
                                             'rank': key_number,
                                             'below only': below,
                                            'tied with': tied_with,
                                            }
                    break
            elif not (fighter in fighters_with_top_5_stats[stat][key]):
                count+=len(fighters_with_top_5_stats[stat][key])
                key_number+=1
                below.extend(fighters_with_top_5_stats[stat][key])
            if count>=250:
                break
    for stat in fighters_with_bottom_5_stats.keys():
        #we count how many fighters rank higher
        count=0
        key_number=1
        above=[]
        for key in fighters_with_bottom_5_stats[stat].keys():
            #print(stat,key,fighters_with_bottom_5_stats[stat][key])
            if fighter in fighters_with_bottom_5_stats[stat][key]:
                if len(fighters_with_bottom_5_stats[stat][key])<250:
                    tied_with = fighters_with_bottom_5_stats[stat][key].copy()
                    tied_with.remove(fighter)
                    interesting_stats[stat]={'value': key,
                                             'rank': key_number,
                                             'above only': above,
                                            'tied with': tied_with,
                                            }
                    break
            elif not fighter in fighters_with_bottom_5_stats[stat][key]:
                count+=len(fighters_with_bottom_5_stats[stat][key])
                key_number+=1
                above.extend(fighters_with_bottom_5_stats[stat][key])
            if count>=250:
                break
    sorted_result = {k: v for k, v in sorted(interesting_stats.items(), key=lambda item: item[1]['rank'])}
    while len(list(sorted_result.keys()))>5:
        key = list(sorted_result.keys())[-1]
        sorted_result.pop(key)
    return sorted_result
        
```
